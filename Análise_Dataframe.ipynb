IMPORTANDO ARQUIVO

```python
import pandas as pd #IMPORTANDO BIBLIOTECA PANDAS

df = pd.read_csv('dados4.csv', index_col='pclass') #lENDO ARQUIVO IMPORTADO
print(df.to_string()) #EXIBINDO TODAS AS MINHAS COLUNAS DO DATAFRAME
```

CODIGO PARA A QUESTAO 01

```python
df.dropna(subset = ['age'], inplace = True)
# O metodo ".dropna", dropa todas as linhas vazias, garantindo assim dados mais consistentes
df['age'] = df['age'].astype(int) # O metodo ".astype" está convertendo minha coluna idade de float para INT, para melhor visualização

moda_da_idade = df['age'].mode()
#Com o metodo ".mode()" determinamos o valor que mais se repete, sendo a moda.
#Em "['idade']" estamos passando a coluna que queremos e extrair a informação.
print(moda_da_idade)

df.to_csv('Resposta01.txt')
#O metodo "to_csv" será o responsável por gerar nosso arquivo
```

CÓDIGO PARA RESPONDER A QUESTÃO 2

```python
sex=df['sex'].value_counts().to_string()
gender=sex.split()
print('2) Número de homens=', gender[1],', Número de mulheres =', gender[3])
print('Somatório entre homens e mulheres=',(len(df.sex)))
```

CÓDIGO PARA RESPONDER A QUESTÃO 3

```python
import matplotlib.pyplot as plt #importando biblioteca Matplot para gráficos

survivors=df['survived'].value_counts().to_string()
survivorsSplit=survivors.split()
print('3)Número de mortos =',survivorsSplit[1], 'Número de sobreviventes =', survivorsSplit[3])
print()

#CÓDIGO DO GRÁFICO DE PIZZA
labels = ['Não-sobreviventes', 'Sobreviventes']
sizes = [809,500] #dados a serem computados pelo gráfico

plt.figure(figsize=(5, 5)) #definição do tamanho do gráfico a ser exibido
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140) #definição da aparência do gráfico de pizza
plt.axis('equal')
plt.title('Número de sobreviventes e mortos')
plt.savefig('Resposta03.png') #gerar um arquivo que mostre o gráfico
plt.show() #comando para exibir o gráfico
print()
```
